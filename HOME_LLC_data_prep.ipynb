```python
# Reading Per Capita GDP Data into a dataframe quarterly data
df_pcgdp = pd.read_csv("/content/Per_Capita_GDP.csv")
df_pcgdp =df_pcgdp[8:92].reset_index(drop=True)
df_pcgdp['quarter'] = pd.DatetimeIndex(df_pcgdp["DATE"]).quarter
df_pcgdp["DATE"] = pd.to_datetime(df_pcgdp["DATE"])
print(df_pcgdp.shape)
df_pcgdp.head()
```

```python
#Reading Housing Inventory Estimate Total Housing Units
df_HIETHU = pd.read_csv("/content/Housing Inventory Estimate Total Housing Units in the United States.csv")
df_HIETHU = df_HIETHU[7:91].reset_index(drop=True)
df_HIETHU["DATE"] = df_pcgdp['DATE']
df_HIETHU['quater']  = pd.DatetimeIndex(df_HIETHU["DATE"]).quarter
df_HIETHU["DATE"] = pd.to_datetime(df_HIETHU["DATE"])
print(df_HIETHU.shape)
df_HIETHU.tail()
```

```python
#Reading Housing_Inventory_Estimate_Vacant_Housing_Units quaterly data
df_HIEVHU = pd.read_csv("/content/Housing_Inventory_Estimate_Vacant_Housing_Units_in_the_United _States.csv")
df_HIEVHU = df_HIEVHU[7:91].reset_index(drop=True)
df_HIEVHU['quater']  = pd.DatetimeIndex(df_HIEVHU["DATE"]).quarter
df_HIEVHU["DATE"] = pd.to_datetime(df_HIEVHU["DATE"])
print(df_HIEVHU.shape)
df_HIEVHU.tail()
```

```python

#Reading Median_Sales_Price_of_Houses_Sold quaterly data
df_Median_sale_price = pd.read_csv("/content/Median_Sales_Price_of_Houses_Sold_for_the_United_States.csv")
df_Median_sale_price  = df_Median_sale_price[8:92].reset_index(drop=True)
df_Median_sale_price['Quater']  = pd.DatetimeIndex(df_Median_sale_price["DATE"]).quarter
df_Median_sale_price['Year']  = pd.DatetimeIndex(df_Median_sale_price["DATE"]).year
df_Median_sale_price["DATE"] = pd.to_datetime(df_Median_sale_price["DATE"])
print(df_Median_sale_price.shape)
df_Median_sale_price.tail()
```

```python
# Concating dataframes having quarterly data to create one dataframe
df2 = pd.DataFrame()
df_byquater = [df_pcgdp,df_GDP ,df_HIEROHU,df_HIETHU ,df_HIEVHU,df_Median_sale_price]
for df1 in df_byquater:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df2 = pd.concat([df2,df1], axis = 1)
column_to_drop = 'Year'
df2.drop(columns=column_to_drop, inplace=True)
print(df2.shape)
df2.head()
```

```python
# Concating df and df2
result_df = pd.concat([df, df2], axis=1)
result_df.head()
```

```python
#Reading Total Household yearly data
df_TTHD = pd.read_csv("/content/TTLHH.csv")
df_TTHD = df_TTHD[2:23].reset_index(drop=True)
df_TTHD["DATE"] = pd.to_datetime(df_TTHD["DATE"])
print(df_TTHD.shape)
df_TTHD.tail()
```

```python
#Reading Urban population percentage yearly data
df_Urban_pop_per = pd.read_csv("/content/API_SP.URB.TOTL.IN.ZS_DS2_en_csv_v2_5996759.csv")
df_Urban_pop_per.rename(columns={'date': 'DATE'}, inplace=True)
df_Urban_pop_per = df_Urban_pop_per[2:].reset_index(drop=True)
df_Urban_pop_per["DATE"] = pd.to_datetime(df_Urban_pop_per["DATE"])
print(df_Urban_pop_per.shape)
df_Urban_pop_per.head()
```

```python
#Reading U.S_population_gender yearly data
df_pop_gender= pd.read_excel("/content/U.S_population_gender.xlsx")
df_pop_gender.rename(columns={'Year': 'DATE'}, inplace=True)
df_pop_gender['DATE'] = df_Urban_pop_per["DATE"].reset_index(drop=True)
df_pop_gender["DATE"] = pd.to_datetime(df_pop_gender["DATE"])
print(df_pop_gender.shape)
df_pop_gender.head()
```

```python
# Concating dataframes having yearly data to create one dataframe
df3 = pd.DataFrame()
df_byquater = [df_pop_gender ,df_TTHD,df_Urban_pop_per]
for df1 in df_byquater:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df3 = pd.concat([df3,df1], axis = 1)
print(df3.shape)
df3.head()
```

```python
# Concating result_df and df3
result_df_final = pd.concat([result_df, df3], axis=1)
result_df_final.head()
```

```python
# Filling missing values in the Per_Capita_GDP column using linear interpolation
result_df_final["A939RX0Q048SBEA"] = result_df_final["A939RX0Q048SBEA"].interpolate()
```

```python
#now lets use linear interpolation on every column which has null values using for loop
null_cols = [ 'quarter', 'GDP', 'quarter',
       'ERNTOCCUSQ176N', 'quater', 'ETOTALUSQ176N', 'quater', 'EVACANTUSQ176N',
       'quater', 'MSPUS', 'Quater', 'Year', 'male', 'female', 'M per 100 F',
       'TTLHH', 'urban_pop_per']
for i in null_cols:
  result_df_final[i] = result_df_final[i].interpolate()
result_df_final.head()
```
